```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t100 = pd.read_csv('simulation_results_100_runs.csv')
t1000 = pd.read_csv('simulation_results_1000_runs.csv')
t10000 = pd.read_csv('simulation_results_10000_runs.csv')
```

```python
plt.hist(t100['newly_infected'], bins=50)
plt.title('100 Trials: Histogram of Newly Infected')
plt.xlabel('Number Newly Infected')
plt.ylabel('Count')
plt.show()

plt.hist(t1000['newly_infected'], bins=50)
plt.title('1000 Trials: Histogram of Newly Infected')
plt.xlabel('Number Newly Infected')
plt.ylabel('Count')
plt.show()

plt.hist(t10000['newly_infected'], bins=50)
plt.title('10,000 Trials: Histogram of Newly Infected')
plt.xlabel('Number Newly Infected')
plt.ylabel('Count')
plt.show()
```

```python
# Get non-zero counts
non_zeros_t100 = t100['newly_infected'].value_counts().drop(0, errors='ignore')
case_percent_t100 = (non_zeros_t100.sum() / 100) * 100
print('Percent of Non-Zero Runs for 100 Trials:', case_percent_t100, '%')

non_zeros_t1000 = t1000['newly_infected'].value_counts().drop(0, errors='ignore')
case_percent_t1000 = (non_zeros_t1000.sum() / 1000) * 100
print('Percent of Non-Zero Runs for 1000 Trials:', case_percent_t1000, '%')

non_zeros_t10000 = t10000['newly_infected'].value_counts().drop(0, errors='ignore')
case_percent_t10000 = (non_zeros_t10000.sum() / 10000) * 100
print('Percent of Non-Zero Runs for 10,000 Trials:', case_percent_t10000, '%')
```

```python
# Filter newly_infected column and create non-zero plots
nz_t100 = t100[t100['newly_infected'] != 0]['newly_infected']

plt.hist(nz_t100, bins=50)
plt.title('100 Trials: Histogram of Newly Infected (Non-Zero Values Only)')
plt.xlabel('Number Newly Infected')
plt.ylabel('Count')
plt.show()

nz_t1000 = t1000[t1000['newly_infected'] != 0]['newly_infected']

plt.hist(nz_t1000, bins=50)
plt.title('1000 Trials: Histogram of Newly Infected (Non-Zero Values Only)')
plt.xlabel('Number Newly Infected')
plt.ylabel('Count')
plt.show()

nz_t10000 = t10000[t10000['newly_infected'] != 0]['newly_infected']

plt.hist(nz_t10000, bins=50)
plt.title('10,000 Trials: Histogram of Newly Infected (Non-Zero Values Only)')
plt.xlabel('Number Newly Infected')
plt.ylabel('Count')
plt.show()
```

```python
# maximum newly infected
max_spread_t100 = t100['newly_infected'].max()
print('Maximum Newly Infected over 48 Hours (100 Trials):',max_spread_t100)

max_spread_t1000 = t1000['newly_infected'].max()
print('Maximum Newly Infected over 48 Hours (1000 Trials):',max_spread_t1000)

max_spread_t10000 = t10000['newly_infected'].max()
print('Maximum Newly Infected over 48 Hours (10,000 Trials):',max_spread_t10000)
print()

# Means and Variance
mean_t100 = t100['newly_infected'].mean()
variance_t100 = t100['newly_infected'].var()
print('Newly Infected (100 Trials): Mean =', mean_t100, ' Variance =', round(variance_t100, 2))

mean_t1000 = t1000['newly_infected'].mean()
variance_t1000 = t1000['newly_infected'].var()
print('Newly Infected (1000 Trials): Mean =', round(mean_t1000, 2), ' Variance =', round(variance_t1000, 2))

mean_t10000 = t10000['newly_infected'].mean()
variance_t10000 = t10000['newly_infected'].var()
print('Newly Infected (10,000 Trials): Mean =', round(mean_t10000, 2), ' Variance =', round(variance_t10000, 2))
```
